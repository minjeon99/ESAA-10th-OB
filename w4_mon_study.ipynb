## **02. 텍스트 사전 준비 작업(텍스트 전처리) - 텍스트 정규화**

텍스트를 입력 데이터로 사용하기 위해 텍스트 정규화 작업은 매우 중요하다.
- 클렌징(Cleansing)
- 토큰화(Tokenization)
- 필터링/스톱 워드 제거/철자 수정
- Stemming
- Lemmatization

### **클렌징**

- 텍스트에서 분석에 방해되는 불필요한 문자, 기호 등을 제거

  예) HTML, XML 태그, 특정 기호 등

### **텍스트 토큰화**

- 문장을 분리하는 문장 토큰화
- 문장에서 단어를 분리하는 단어 토큰화

#### **문장 토큰화**

- 문장의 마침표(.), 개행문자(/n) 등 문장의 마지막을 뜻하는 기호에 따라 분리하는 것이 일반적
- 정규 표현식에 따른 문장 토큰화도 가능
- NLTK의 `sent_tokenize` 이용

```python
from nltk import sent_tokenize
import nltk
nltk.download('punkt') # 마침표, 개행 문자 등의 데이터 세트 다운로드

text_sample = 'The Matrix is everywhere its all around us, here even in this room. \
               You can see it out your window or on your television. \
               You feel it when you go to work, or go to church or pay your taxes.'

sentences = sent_tokenize(text=text_sample)
```

```python
print(type(sentences), len(sentences))
print(sentences)
```

- 각각의 문장으로 구성된 list 객체 반환

#### **단어 토큰화**

- 공백, 콤마(,), 마침표(.), 개행문자 등으로 단어를 분리하지만, 정규 표현식을 이용해 다양한 유형으로 토큰화 수행 가능
- Bag of Word와 같이 단어의 순서가 중요하지 않은 경우 단어 토큰화만 사용해도 충분
  - 문장 토큰화는 각 문장이 가지는 시맨틱적 의미가 중요한 요소로 사용될 때 사용
- NLTK의 `word_tokenize()` 이용

```python
from nltk import word_tokenize

sentence = 'The Matrix is everywhere its all around us, here even in this room.'
words = word_tokenize(sentence)
print(type(words), len(words))
print(words)
```

sent_tokenize, word_tokenize 조합해 문서에 대한 모든 단어 토큰화

```python
from nltk import word_tokenize, sent_tokenize

# 여러 개의 문장으로 된 입력 데이터를 문장별로 단어 토큰화하는 함수
def tokenize_text(text):
  # 문장별로 분리 토큰
  sentences = sent_tokenize(text)
  # 분리된 문장별 단어 토큰화
  word_tokens = [word_tokenize(sentence) for sentence in sentences]
  return word_tokens

# 여러 문장에 대해 문장별 단어 토큰화
word_tokens = tokenize_text(text_sample)
print(type(word_tokens), len(word_tokens))
print(word_tokens)
```

- 문장을 단어별로 하나씩 토큰화할 경우 문맥적 의미는 무시됨
- **n-gram**: 연속된 n개의 단어를 하나의 토큰화 단위로 분리

#### **스톱 워드 제거**

- Stop word: 분석에 큰 의미가 없는 단어
  - 영어의 is, the, a, will 등
  - 문법적 특성으로 인해 빈번하게 나타나므로, 사전에 제거하지 않으면 오히려 중요한 단어로 인지될 수 있음
- 언어별 스톱 워드 목록화
  - NLTK의 stopwords에서 제공

```python
import nltk
nltk.download('stopwords')
```

```python
print('영어 stop words 개수:', len(nltk.corpus.stopwords.words('english')))
print(nltk.corpus.stopwords.words('english')[:20])
```

예제에서 생성된 word_tokens 리스트에서 stopwords를 필터링으로 제거하여 의미있는 단어만 추출

```python
import nltk

stopwords = nltk.corpus.stopwords.words('english')
all_tokens = []
# 위 예제에서 3개의 문장별로 얻은 word_tokens list에 대해 스톱 워드 제거
for sentence in word_tokens:
  filtered_words = []
  # 개별 문장별로 토큰화된 문장 list에 대해 스톱 워드 제거
  for word in sentence:
    # 모두 소문자로 변환
    word = word.lower()
    # 토큰화된 개별 단어가 스톱 워드에 포함되지 않으면 word_tokens에 추가
    if word not in stopwords:
      filtered_words.append(word)
  all_tokens.append(filtered_words)

print(all_tokens)
```

#### **Stemming과 Lemmatization**

- 영어의 경우 과거/현재, 3인칭 단수 여부, 진행형 등 많은 조건에 따라 원래 단어가 변화
- Stemming, Lemmatization: 문법적, 의미적으로 변화하는 단어의 원형 찾기
  - 두 기능 모두 목적은 유사하지만, Lemmatization이 더 정교하며 의미론적 기반에서 원형 찾기
  - Stemming은 일반적인 방법, 단순화된 방법을 적용해 일부 철자가 훼손된 어근 단어를 추출하는 경향
  - Lemmatization: 품사와 같은 문법적 요소, 의미적 부분을 감안해 정확한 철자로 된 어근 단어 찾기
    - 더 오랜 시간 소요
- NLTK의 다양한 Stemmer: Porter, Lancaster, Snowball Stemmer
- NLTK의 Lemmatization: WordNetLemmatizer

```python
from nltk.stem import LancasterStemmer
stemmer = LancasterStemmer()

print(stemmer.stem('working'), stemmer.stem('works'), stemmer.stem('worked'))
print(stemmer.stem('amusing'), stemmer.stem('amuses'), stemmer.stem('amused'))
print(stemmer.stem('happier'), stemmer.stem('happiest'))
print(stemmer.stem('fancier'), stemmer.stem('fanciest'))
```

- work: 기본 단어 work에 ing, s, ed가 붙는 단순한 변화이므로 제대로 인식
- amuse: amus에 ing, s, ed가 붙으므로 정확한 단어 인식 X
- happy, fancy도 변형된 단어의 정확한 원형 찾지 X

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')

lemma = WordNetLemmatizer()
print(lemma.lemmatize('amusing', 'v'), lemma.lemmatize('amuses', 'v'), lemma.lemmatize('amused', 'v'))
print(lemma.lemmatize('happier', 'a'), lemma.lemmatize('happiest', 'a'))
print(lemma.lemmatize('fancier', 'a'), lemma.lemmatize('fanciest', 'a'))
```

- Stemmer보다 정확하게 원형 단어 추출

## **03. Bag of Words - BOW**

문서가 가지는 모든 단어를 문맥, 순서를 무시하고 일괄적으로 단어에 빈도 값을 부여해 피처 값 추출하는 모델
1. 문장의 모든 단어에서 중복 제거하고 각 단어(feature or term)를 칼럼 형태로 나열. 각 단어에 고유의 인덱스 부여
2. 개별 문장에서 해당 단어가 나타나는 횟수(Occurrence)를 각 단어(단어 인덱스)에 기재
- 장점: 쉽고 빠른 구축
  - 단순히 단어 발생 횟수에 기반하지만, 예상보다 문서의 특징을 잘 나타낼 수 있는 모델
  - 전통적으로 여러 분야에서 활용
- 단점
  - 문맥 의미(Semantic Context) 반영 부족: 단어의 순서 고려 X -> 문장 내에서 단어의 문맥적 의미 무시
    - 보완하기 위한 n_gram 기법이 있지만, 제한적인 부분에 그침
  - 희소 행렬(Sparse Matrix) 문제(희소성, 희소 행렬): 희소 행렬 형태의 데이터 세트가 만들어지기 쉬움. 매우 많은 단어가 칼럼으로 만들어짐.
    - 대부분의 값이 0으로 채워지는 형태

    <-> 밀집 행렬(Dense Matrix)


### **BOW 피처 벡터화**

- 피처 벡터화: 텍스트를 특정 의미를 가지는 숫자형 값인 벡터 값으로 변환하는 것
  - 기존 텍스트 데이터를 또 다른 형태의 피처의 조합으로 변경하므로 넓은 범위의 피처 추출에 포함
- BOW 모델에서의 피처 벡터화
  - 모든 문서의 모든 단어를 칼럼 형태로 나열
  
    -> 각 문서에서 해당 단어의 횟수나 정규화된 빈도를 값으로 부여하는 데이터 세트 모델로 변경
  - M개의 문서, 전체 문서에 N개의 단어 => M x N개의 단어 피처로 이루어진 행렬
- BOW의 피처 벡터화의 방식
  1. **카운트 기반**의 벡터화: 각 문서에서 해당 단어가 나타나는 횟수(Count) 부어
    - 카운트 값이 높을수록 중요 단어로 인식
    - 언어의 특성상 문장에서 자주 사용될 수밖에 없는 단어까지 높은 값 부여하는 문제
  2. **TF-IDF(Term Frequency - Inver Document Frequency)** 기반의 벡터화
    - 개별 문서에서 자주 나타나는 단어에 높은 가중치,       모든 문서에서 자주 나타나는 단어에 페널티
    - 문서의 텍스트가 길고, 문서 개수가 많은 경우 TF-IDF가 더 좋은 예측 성능 보장

#### **BOW 벡터화를 위한 희소 행렬**

- BOW 형태를 가진 언어 모델의 피처 벡터화는 대부분 희소 행렬
- 희소 행렬은 너무 많은 불필요한 0 값이 메모리 공간에 할당되어 메모리 공간 많이 필요
- 행렬 크기가 커서 연산 시 데이터 엑세스를 위한 많은 시간 소요
- 희소 행렬이 물리적으로 적은 메모리 공간을 차지할 수 있도록 변환하는 방식 - CSR, COO 방식
  - 사이킷런의 CountVectorizer/TfidVectorizer: CSR 형태의 희소 행렬 반환

#### **희소 행렬 - COO 형식**

- COO(Coordicate: 좌표): 0이 아닌 데이터만 별도의 데이터 배열(Array)에 저장, 데이터가 가리키는 행/열 위치를 별도의 배열로 저장
  - 파이썬에서는 주로 Scipy 이용 - Scipy의 `coo_matrix` 클래스

```python
import numpy as np
dense = np.array([ [3,0,1], [0,2,0]])
```

```python
from scipy import sparse

# 0이 아닌 데이터 추출
data = np.array([3,1,2])

# 행 위치와 열 위치를 각각 배열로 생성
row_pos = np.array([0,0,1])
col_pos = np.array([0,2,1])

# sparse 패키지의 coo_matrix를 이용해 COO 형식으로 희소 행렬 생성
sparse_coo = sparse.coo_matrix((data, (row_pos, col_pos)))
```

```python
# 밀집 행렬로 출력
sparse_coo.toarray()
```

#### **희소 행렬 - CSR 형식**

- COO 형식: 행/열 위치 표현을 위해 반복적인 위치 데이터 사용 문제
  - CSR(Compressed Sparse Row)가 해결

  image.png

- 순차적인 같은 값이 반복적으로 나타남
- 행 위치 배열이 0부터 순차적으로 증가하는 값으로 이루어진 특성 고려 => 행 위치 배열의 고유한 값의 시작 위치만 표기하기

image.png

```python
from scipy import sparse

dense2 = np.array([[0,0,1,0,0,5],
             [1,4,0,3,2,5],
             [0,6,0,3,0,0],
             [2,0,0,0,0,0],
             [0,0,0,7,0,8],
             [1,0,0,0,0,0]])

# 0이 아닌 데이터 추출
data2 = np.array([1, 5, 1, 4, 3, 2, 5, 6, 3, 2, 7, 8, 1])

# 행 위치와 열 위치를 각각 array로 생성
row_pos = np.array([0, 0, 1, 1, 1, 1, 1, 2, 2, 3, 4, 4, 5])
col_pos = np.array([2, 5, 0, 1, 3, 4, 5, 1, 3, 0, 3, 5, 0])

# COO 형식으로 변환
sparse_coo = sparse.coo_matrix((data2, (row_pos, col_pos)))

# 행 위치 배열의 고유한 값의 시작 위치 인덱스를 배열로 생성
row_pos_ind = np.array([0,2,7,9,10,12,13])

# CSR 형식으로 변환
sparse_csr = sparse.csr_matrix((data2, col_pos, row_pos_ind))

print('COO 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_coo.toarray())
print('CSR 변환된 데이터가 제대로 되었는지 다시 Dense로 출력 확인')
print(sparse_csr.toarray())
```

- 행 위치 배열 만들 때 마지막 인덱스도 포함해야 함(끝이 어디인지 알 수 있게)
